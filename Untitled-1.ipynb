```python
#W5A23
def quadratic_roots_count(a, b, c):
    if a == 0:
        return 1 if b != 0 else 0
    delta = b*b - 4*a*c
    if delta > 0:
        return 2
    elif delta == 0:
        return 1
    else:
        return 0
```

```python
#W5A25
def area(x1,y1,x2,y2,x3,y3):
    return abs((x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)) / 2)

def point_in_quadrilateral(quad, p):
    (x, y) = p
    A, B, C, D = quad
    
    total = (area(*A, *B, *C) +
             area(*A, *C, *D))
    
    s = (area(x,y,*A,*B) + area(x,y,*B,*C) +
         area(x,y,*C,*D) + area(x,y,*D,*A))
    
    return abs(total - s) < 1e-6
```

```python
#W5A28
import math

def angle_ABC(x1,y1,x2,y2,x3,y3):
    BA = (x1-x2, y1-y2)
    BC = (x3-x2, y3-y2)
    
    dot = BA[0]*BC[0] + BA[1]*BC[1]
    lenBA = math.sqrt(BA[0]**2 + BA[1]**2)
    lenBC = math.sqrt(BC[0]**2 + BC[1]**2)
    
    cosA = dot / (lenBA * lenBC)
    return math.degrees(math.acos(cosA))
```

```python
#W5A33
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    fare = 0
    
    # giá mở cửa
    fare += 10000
    
    # giá km
    if km <= 1:
        fare += 0
    elif km <= 10:
        fare += (km-1) * 12000
    else:
        fare += 9 * 12000 + (km-10) * 10000
    
    # phí chờ
    fare += wait_minutes * 500
    
    # ban đêm
    if start_hour >= 22 or start_hour <= 5:
        fare *= 1.2
    
    return int(fare)
```

```python
#W5A34
def is_unique_digits(x):
    s = str(x)
    return len(set(s)) == len(s)

def sum_unique_digits(n):
    total = 0
    for i in range(1, n+1):
        if is_unique_digits(i):
            total += i
    return total
```
